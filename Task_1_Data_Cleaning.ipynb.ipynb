```python
#import the pandas library
import pandas as pd
```

```python

#read/load the data sets as ds
ds = pd.read_csv('raw data.csv')

```

```python
ds
```

```python

#check the number of rows and columns
print(f'we have {ds.shape[0]} rows and {ds.shape[1]} columns')

```

```python
#standardise incosistent data in categorical collumns formats i.e remove accidental white spaces and standardize lowercase

ds['customer_name'] = ds['customer_name'].str.strip().str.lower()
ds['product'] = ds['product'].str.strip().str.lower()
ds['category'] = ds['category'].str.strip().str.lower()
```

```python
#check for  missing values
print(ds.isnull().sum()) 

```

```python
# filling of the missing values
ds['customer_name'] = ds['customer_name'].fillna('unknown')
ds['quantity'] = ds['quantity'].fillna(ds['quantity'].median())
ds['price'] = ds['price'].fillna(ds['price'].median())
```

```python
ds
```

```python

#check for duplicate values
print(ds.duplicated().sum())
```

```python

#remove duplicate values

ds = ds.drop_duplicates()
```

```python
ds
```

```python
ds.to_csv("cleaned_data.csv" , index=True) 
```
